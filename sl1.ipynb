```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sigmoid():
    x = np.linspace(-10, 10, 100)  # Generate 100 equally spaced values from -10 to 10
    y = 1 / (1 + np.exp(-x))  # Compute the sigmoid function values
    
    plt.plot(x, y)
    plt.xlabel('Input')
    plt.ylabel('Sigmoid Output')
    plt.title('Sigmoid Activation Function')
    plt.grid(True)
    plt.show()
plot_sigmoid()    

```

```python
def plot_tanh():
    # Generate values for x
    x = np.linspace(-10, 10, 100)

    # Compute tanh values for corresponding x
    tanh = np.tanh(x)

    # Plot the tanh function
    plt.plot(x, tanh)
    plt.title("Hyperbolic Tangent (tanh) Activation Function")
    plt.xlabel("x")
    plt.ylabel("tanh(x)")
    plt.grid(True)
    plt.show()
plot_tanh()   
```

```python
def plot_relu():
    # Generate values for x
    x = np.linspace(-10, 10, 100)

    # Compute ReLU values for corresponding x
    relu = np.maximum(0, x)

    # Plot the ReLU function
    plt.plot(x, relu)
    plt.title("ReLU Activation Function")
    plt.xlabel("x")
    plt.ylabel("ReLU(x)")
    plt.grid(True)
    plt.show()
plot_relu()    
```

```python
import pandas as pd
# Identity activation function
def identity_activation(x):
    return x
inputs = pd.DataFrame({'input': range(-5, 6)}) 
inputs['output'] = inputs['input'].apply(identity_activation)
plt.figure(figsize=(8, 6))
plt.plot(inputs['input'], inputs['output'], label="Identity Function", color='blue')
plt.title('Identity Activation Function', fontsize=14)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Output', fontsize=12)
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)  # Create a range of values for the x-axis
y = np.exp(x) / np.sum(np.exp(x))  # Compute the softmax-like activation

# Plotting the softmax
plt.plot(x, y, label='Softmax')
plt.xlabel('Input')
plt.ylabel('Activation')
plt.title('Activation Functions')
plt.legend()
plt.grid(True)
plt.show()
```
